```python
img = cv2.imread(r"C:\Users\D. RHUTHVIK\Desktop\Final Year Project\all imgs\abdomen.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
img2 = img.reshape(-1)
df = pd.DataFrame()
df['Original Image'] = img2

```

```python
num = 1  #To count numbers up in order to give Gabor features a lable in the data frame
kernels = []
for theta in range(4):
    theta = theta / 4. * np.pi
    for sigma in (1, 3,5):
        for lamda in np.arange(0, np.pi, np.pi / 4):
            for gamma in (0.05, 0.5):
                gabor_label = 'Gabor' + str(num)
                ksize=9
                kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, 0, ktype=cv2.CV_32F)
                kernels.append(kernel)
                fimg = cv2.filter2D(img, cv2.CV_8UC3, kernel)
                filtered_img = fimg.reshape(-1)
                cv2.imwrite("C:/Users/D. RHUTHVIK/Desktop/Final Year Project/all imgs/abdomen2/"+gabor_label+".png", filtered_img.reshape(img.shape))
                df[gabor_label] = filtered_img  #Labels columns as Gabor1, Gabor2, etc.
                print(gabor_label, ': theta=', theta, ': sigma=', sigma, ': lamda=', lamda, ': gamma=', gamma)
                num += 1  #Increment for gabor column label
print(df.head())

                    
```

```python
def showimage(myimage, figsize=[10,10]):
    if (myimage.ndim>2):  #This only applies to RGB or RGBA images (e.g. not to Black and White images)
        myimage = myimage[:,:,::-1] #OpenCV follows BGR order, while matplotlib likely follows RGB order
         
    fig, ax = plt.subplots(figsize=figsize)
    ax.imshow(myimage, cmap = 'gray', interpolation = 'bicubic')
    plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
    plt.show()
```

```python
import cv2
import os

# Define the folder path containing the images
folder_path = r'C:\Users\D. RHUTHVIK\Desktop\Final Year Project\all imgs\abdomen2'

# Loop through all the files in the folder
for filename in os.listdir(folder_path):
    # Check if the file is an image
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # Read the image
        image = cv2.imread(os.path.join(folder_path, filename))
        
        # Convert the image to grayscale
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # Perform Canny edge detection on the image
        edges = cv2.Canny(gray, 100, 200)
        
        # Create a named window and set its size
        cv2.namedWindow(filename + ' - Original Image', cv2.WINDOW_NORMAL)
        cv2.resizeWindow(filename + ' - Original Image', 600, 600)
        cv2.namedWindow(filename + ' - Canny Edges', cv2.WINDOW_NORMAL)
        cv2.resizeWindow(filename + ' - Canny Edges', 600, 600)
        
        # Display the original image and the edges
        cv2.imshow(filename + ' - Original Image', image)
        cv2.imshow(filename + ' - Canny Edges', edges)
        cv2.waitKey(0)  # Wait for user to press any key
        
cv2.destroyAllWindows()  # Close all windows

```
